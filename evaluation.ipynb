```python
images = read_in_data('./Clusters_2D_100.txt')
images_30 = read_in_data('./Clusters_2D_100_30.txt')
```

```python
'''
algorithms: 'local maxima', 'local maxima denoised', 'local maxima denoised double', 
'hr local maxima denoised', 'hr local maxima denoised double'
'''

# 3*3 neighborhood
found_1 = find_points(images, 'local maxima denoised', half_size=1)
result1 = judge(images, found_1, threshold_error=0.25)
# 3*3 neighborhood
found_30_1 = find_points(images_30, 'local maxima denoised', half_size=1)
result2 = judge(images_30, found_30_1, threshold_error=0.25)
compare_results([result1, result2], ['10 Clusters, 3*3 Neighborhood', '30 Clusters, 3*3 Neighborhood'])
```

```python
'''
algorithms: 'local maxima', 'local maxima denoised', 'local maxima denoised double', 
'hr local maxima denoised', 'hr local maxima denoised double'
'''

# 5*5 neighborhood
found_2 = find_points(images, 'local maxima denoised', half_size=2)
result3 = judge(images, found_2, threshold_error=0.25)
# 5*5 neighborhood
found_30_2 = find_points(images_30, 'local maxima denoised', half_size=2)
result4 = judge(images_30, found_30_2, threshold_error=0.25)
compare_results([result3, result4], ['10 Clusters, 5*5 Neighborhood', '30 Clusters, 5*5 Neighborhood'])
```

```python
compare_results([result1, result3], ['10 Clusters, 3*3 Neighborhood', '10 Clusters, 5*5 Neighborhood'])
compare_results([result2, result4], ['30 Clusters, 3*3 Neighborhood', '30 Clusters, 5*5 Neighborhood'])
```
